```python
import pandas as pd
import RAG_MTG_Cards as RMTG
from tqdm import tqdm
```

```python
test_files = pd.read_csv('RAG_MTG_Test.csv')

rag_output_values = []
base_output_values = []

for card_name in tqdm(test_files['Card Name'], desc='Processing'):
    rag_output = RMTG.rag_query(card_name, RAG=True)
    rag_output_values.append(rag_output)
    
    base_output = RMTG.rag_query(card_name, RAG=False)
    base_output_values.append(base_output)

test_files['RAG Output'] = rag_output_values
test_files['Base Output'] = base_output_values
```

```python
test_files
```

```python
raw_df = pd.read_csv("raw/filtered_oracle_database.csv")

original_values = []

for card_name in tqdm(test_files['Card Name'], desc='Processing'):

    original_row = raw_df[raw_df['name'] == card_name]
    
    formatted_row = ""
    for column_name, value in original_row.items():
        formatted_row += f"{column_name}: {value.values[0]}\n"
    original_values.append(formatted_row)

test_files['Original Text'] = original_values
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)

def calculate_rouge_l(row, reference_column, hypothesis_column):
    reference = row[reference_column]
    hypothesis = row[hypothesis_column]
    rouge_score_result = scorer.score(reference, hypothesis)
    return rouge_score_result['rougeL'].fmeasure

# Calculate ROUGE-L score and store in 'RAG-score' column
test_files['RAG-score'] = test_files.apply(calculate_rouge_l, axis=1)
test_files
```
